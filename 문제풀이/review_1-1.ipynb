문제3) 기초 통계량 (평균, 중앙값, IQR, 표준편차 등)    
- mpg의 평균값, 중앙값, 최빈값을 구하여라 
- mpg의 분산, 표준편차, IQR, 최소값, 최대값을 구하여라 
- mpg의 첨도, 왜도를 구하여라 

```python
mean_mpg = mtcars['mpg'].mean()
median_mpg = mtcars['mpg'].median()
mode_mpg = mtcars['mpg'].mode()

var_mpg = mtcars['mpg'].var()
std_mpg = mtcars['mpg'].std()

q1 = mtcars['mpg'].quantile(0.25)
q3 = mtcars['mpg'].quantile(0.75)
iqr_mpg = abs(q1-q3)

min_mpg = mtcars['mpg'].min()
max_mpg = mtcars['mpg'].max()

skew_mpg = mtcars['mpg'].skew()
kurt_mpg = mtcars['mpg'].kurt()

print('mpg 평균 : ', mean_mpg)
print('mpg 중앙값 : ', median_mpg)
print('mpg 최빈값 : ', mode_mpg)
print('mpg 분산 : ', var_mpg)
print('mpg 표준편차 : ', std_mpg)
print('mpg IQR : ', iqr_mpg)
print('mpg 최소값 : ', min_mpg)
print('mpg 최대값 : ', max_mpg)
print('mpg 왜도 : ', skew_mpg)
print('mpg 첨도 : ', kurt_mpg)
```

# IQR 이상치 처리 
## upper = Q3+1.5*IQR
## lower = Q1-1.5*IQR

```python

titanic_iqr = titanic.copy()
q1 = titanic_iqr['age'].quantile(0.25)
q3 = titanic_iqr['age'].quantile(0.75)
iqr = abs(q1-q3)
cond1 = titanic_iqr['age']< (q3+1.5*iqr)
cond2 = titanic_iqr['age']> (q1-1.5*iqr)

titanic_iqr = titanic_iqr[cond1&cond2]
```

```python
len(titanic_iqr)
```
